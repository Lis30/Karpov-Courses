Импортируйте библиотеку pandas как pd. <br> Загрузите датасет bookings.csv с разделителем ;. <br>Проверьте размер таблицы, типы переменных, а
затем выведите первые 7 строк, чтобы посмотреть на данные.

```python
book = pd.read_csv('bookings.csv', sep = ';')
```

```python
book.info()
```

```python
book.head(7)
```

Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.


```python
book.columns = book.columns.str.lower().str.replace(' ', '_')
book.head(5)
```

Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.


```python
book.query('is_canceled==0').\
          groupby('country').\
          agg({'lead_time': 'count'}).\
          sort_values('lead_time', ascending=False).\
          head(5)
```

На сколько ночей в среднем бронируют отели разных типов?


```python
book.query('is_canceled==0').\
            groupby('hotel').\
            agg({'stays_total_nights': 'mean'}).\
            round(0)
```

Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). <br>
Такое может произойти, например, по причине овербукинга. Сколько подобных наблюдений встретилось в датасете?


```python
book.query('is_canceled == 0').\
                        query('assigned_room_type!=reserved_room_type').\
                        agg({'is_canceled':'count'})
```

Проанализируйте даты запланированного прибытия.<br>
– На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017? <br>


```python
book.query('is_canceled == 0').\
                    query('arrival_date_year == 2016').\
                    groupby('arrival_date_month').\
                    agg({'is_canceled':'count'}).\
                    sort_values('is_canceled', ascending=False).head(1)
```

```python
book.query('is_canceled == 0').\
                        query('arrival_date_year == 2017').\
                        groupby('arrival_date_month').\
                        agg({'is_canceled':'count'}).\
                        sort_values('is_canceled', ascending=False).head(1)
```

– Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

```python
book.query('is_canceled == 1').\
                    query('hotel == "City Hotel"').\
                    groupby(['arrival_date_year', 'arrival_date_month']).\
                    agg({'is_canceled':'count'})
```

Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
book[['adults', 'children', 'babies']].mean()
```

Создайте колонку total_kids, объединив children и babies. Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?


```python
book['total_kids'] = book.children + book.babies
book.query('is_canceled == 0').\
            groupby('hotel').\
            agg({'total_kids': 'mean'})
```

Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), <br>
в противном случае – False. <br> Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах
(churn rate).<br> Укажите, среди какой группы показатель выше.

```python
book['has_kids'] = (book.total_kids>0)
```

```python
#процент отказов к общему количеству
persent_canceled = book.is_canceled.sum() / book.agg({'is_canceled': 'count'})
persent_canceled = round(persent_canceled*100, 2)
persent_canceled
```

```python
#процент отказов от семей с детьми
persent_canceled_family = book.query('total_kids > 0').\
                            is_canceled.sum() / book.query('total_kids > 0').\
                            agg({'is_canceled':'count'})
persent_canceled_family = round(persent_canceled_family*100, 2)
persent_canceled_family
```

```python
#процент отказов от людей без детей
persent_canceled_single = book.query('total_kids == 0').\
                            is_canceled.sum() / book.query('total_kids ==0').\
                            agg({'is_canceled':'count'})
persent_canceled_single = round(persent_canceled_single*100, 2)
persent_canceled_single
```
